```python
full_dot = '●'
empty_dot = '○'

def create_character(character_name, strength, intelligence, charisma):
    # name validation
    if not isinstance(character_name, str):
        return 'The character name should be a string'
    if character_name == '':
        return 'The character should have a name'
    if len(character_name) > 10:
        return 'The character name is too long'
    if ' ' in character_name:
        return 'The character name should not contain spaces'

    # stats validations
    stats_type_check = isinstance(strength, int) and isinstance(intelligence, int) and isinstance(charisma, int)
    if not stats_type_check:
        return 'All stats should be integers'

    stats_min_check = (strength < 1) or (intelligence < 1) or (charisma < 1)
    if stats_min_check:
        return 'All stats should be no less than 1'

    stats_max_check = (strength > 4) or (intelligence > 4) or (charisma > 4)
    if stats_max_check:
        return 'All stats should be no more than 4'

    stats_total_value = strength + intelligence + charisma
    if stats_total_value != 7:
        return 'The character should start with 7 points'

    # main function body
    output = character_name + '\n'
    output += 'STR ' + (full_dot * strength) + (empty_dot * (10-strength)) + '\n'
    output += 'INT ' + (full_dot * intelligence) + (empty_dot * (10-intelligence)) + '\n'
    output += 'CHA ' + (full_dot * charisma) + (empty_dot * (10-charisma))

    return output
```

```python
expected = 'ren\nSTR ●●●●○○○○○○\nINT ●●○○○○○○○○\nCHA ●○○○○○○○○○'
actual = create_character('ren', 4, 2, 1)

if expected == actual:
    print('Good')
else:
    print('Something wrong!')

expected
```
